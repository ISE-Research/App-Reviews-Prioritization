```
import pandas as pd
from sentistrength import PySentiStr
from nltk.tokenize import sent_tokenize
import nltk
nltk.download('punkt')
senti = PySentiStr()
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
path1='/content/drive/My Drive/SentiStrength.jar'
path2='/content/drive/My Drive/SentStrengthData/'
senti.setSentiStrengthPath(path1)
senti.setSentiStrengthLanguageFolderPath(path2)
# senti.setSentiStrengthPath('C:/Users/Forough/PycharmProjects/untitled2/SentiStrength.jar')
# senti.setSentiStrengthLanguageFolderPath('C:/Users/Forough/PycharmProjects/untitled2/SentStrengthData/')
```

```
# path3 = '/content/drive/My Drive/preproceesed_data_of_thesis_part1_final_preprocess_v1_first_section_0_to_19999_reviews.csv'
path3 = '/content/drive/My Drive/preproceesed_data_of_thesis_part2_final_preprocess_v1_second_section_25000_to_50000_reviews.csv'
df = pd.read_csv(path3, encoding= 'unicode_escape')
```

```
list_of_review_sentiments=[]
for column in df[['reviews']]:
  text=df[column].values
  for index in range(len(df)):
    list_of_sentence_tokens=[]
    list_of_sentence_tokens=sent_tokenize(str(text[index]))
    max=-10
    for t in range(len(list_of_sentence_tokens)):
        sentence_sentiment_score=senti.getSentiment(list_of_sentence_tokens[t], score='scale')[0]
        if max < sentence_sentiment_score:
            max=sentence_sentiment_score
    list_of_review_sentiments.append(max)
```

```
df['review_sentiment']=list_of_review_sentiments
```
