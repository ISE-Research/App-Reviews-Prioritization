```
path = '/content/drive/My Drive/merged_data_ready_to_train with intention without feature name.csv'
data = pd.read_csv(path)
```

```
data
```

```
data=data.drop(['f8', 'f9','f10','f11','f13','f15','f16','f17','f18','f19','f20','f22','f23','f24','f25','f26','f27','f28'], axis = 1)
```

```
# Get names of indexes for which column Age has value 30
indexNames = data[ data['output'] == -1].index
# print(indexNames)
# Delete these row indexes from dataFrame
data.drop(indexNames , inplace=True)
data
```

```
# data.loc[(data["output"] ==0)] = 1
data.loc[data.output==0, 'output'] = 1
```

```
# data.loc[(data["output"] ==2)] = 1
data.loc[data.output==2, 'output'] = 1
```

```
# data.loc[(data["output"] ==3)] = 1
data.loc[data.output==3, 'output'] = 1
```

```
# data.loc[(data["output"] >3)] = 0
data.loc[data.output>3, 'output'] = 0
```

```
X = data.drop('output', axis=1)
y = data['output']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

#implementing Kernel SVM

```
svclassifier_poly = SVC(kernel='poly', degree=1)
svclassifier_poly.fit(X_train, y_train)
```

```
y_pred_svclassifier_poly = svclassifier_poly.predict(X_test)
```

```
print(confusion_matrix(y_test, y_pred_svclassifier_poly))
print(classification_report(y_test, y_pred_svclassifier_poly))
```

#implementing  Gaussian Kernel

```
svclassifier_rbf = SVC(kernel='rbf')
svclassifier_rbf.fit(X_train, y_train)
```

```
y_pred_svclassifier_rbf = svclassifier_rbf.predict(X_test)
```

```
print(confusion_matrix(y_test, y_pred_svclassifier_rbf))
print(classification_report(y_test, y_pred_svclassifier_rbf))  
```

#implementing Sigmoid Kernel

```
svclassifier_sigmoid = SVC(kernel='sigmoid')
svclassifier_sigmoid.fit(X_train, y_train)
```

```
y_pred_svclassifier_sigmoid = svclassifier_sigmoid.predict(X_test)
```

```
print(confusion_matrix(y_test, y_pred_svclassifier_sigmoid))
print(classification_report(y_test, y_pred_svclassifier_sigmoid))
```
